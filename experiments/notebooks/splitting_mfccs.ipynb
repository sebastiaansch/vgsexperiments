```python
import torch
import tables
```

```python
def iterate_data(h5_file):
    for x in h5_file.root:
        yield x
```

```python
path ="/Users/sebastiaanscholten/Documents/speech2image-master/experiments/Generating_Flickrwords_mfcc/mfcc/words_mfcc_features.h5"

mfccfile = tables.open_file(path, mode="r+")
mfccfile = [node for node in iterate_data(mfccfile)]
```

```python
from pathlib import Path
data_loc = "output_file/mfcc_features_lefttoright.h5"

if not Path(data_loc).is_file():
    # create h5 output file for preprocessed images and audio
    output_file = tables.open_file(data_loc, mode='a')
    # create the h5 file to hold all image and audio features. This will fail if they already excist such
    # as when you run this file to append new features to an excisting feature file
    for x in mfccfile:
        try:
            # one group for each image file which will contain its vgg16 features and audio captions
            output_file.create_group("/", x._v_name)
        except:
            continue
# else load an existing file to append new features to
else:
    output_file = tables.open_file(data_loc, mode='a')
    
node_list = output_file.root._f_list_nodes()
```

```python
for node in node_list:
    print(node)
```

```python
def left_to_right_mfccsplit(read_file,output_file,append):
#     for node in node_list:
#         print(f"working on node {node}")
        for mfccdata,node in zip(read_file,node_list):
            print(f"Running for {mfccdata}")
            for percent in range(100,0,-1):
                mfcc_vectors = mfccdata.mfcc._f_list_nodes()[0]
                mfcc_columns = mfcc_vectors.shape[0]

                which_column = mfcc_columns - max(1,round(mfcc_columns*(percent/100)))
                mfcc_split = mfcc_vectors[which_column:,:]
                mfcc_node = output_file.create_group(node, append + str(percent))
                mfcc_array = output_file.create_array(mfcc_node,node._v_name,mfcc_split)
```

```python
append = "percent_"
left_to_right_mfccsplit(mfccfile,output_file,append)
```
